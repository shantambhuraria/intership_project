## Exploring the compliance summary data: m2_summarycompliance_jan - jul

#### creating final dataframe for m2_summarycompliance_jan - jul exploration

```python
# Define your query params
compliance_data_index_list =['m2_summarycompliance_jan', 'm2_summarycompliance_feb', 'm2_summarycompliance_mar', 
           'm2_summarycompliance_apr', 'm2_summarycompliance_may', 'm2_summarycompliance_jun', 
           'm2_summarycompliance_jul']

df = None
for index in compliance_data_index_list:
    if df is None:
        df = eu.query(indx=index)
    else:
        df = pd.concat([df, eu.query(indx=index)], ignore_index=True)


df
```

```python
df = df[df["subject"].isin(["022", "223", "012", "231", "122", "147", "245", "061", "314", "127"])]
df
```

```python
# scatterplot for the survey data whether the patient has uploaded the data or not.
sns.scatterplot(data=df, x=df["datetime"], y=df["subject"], hue=df["has_uploaded"])
```

```python
# finding whether the patient is there on site when study is conducted
set(df[['study', 'site']].apply(tuple, axis=1))
```

```python
# scatterplot for the survey data 'study', 'site'
sns.scatterplot(data=df, x=df["datetime"], y=df["subject"], hue=df[['study', 'site']].apply(tuple, axis=1))
```
